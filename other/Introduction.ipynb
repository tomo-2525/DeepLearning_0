## numpyの軽い復習

```python
import numpy as np
```

```python
A = np.array([[1, 2],[3, 4]])
print(A)
```

```python
B = np.array([[3, 0], [0, 6]])
print(B)
```

```python
A + B
```

```python
A * B
```

```python
X = np.array([[51, 55], [14, 19], [0, 4]])
print(X)
```

```python
X[0] #0行目
```

```python
X[: ,1] #1列目を一行で
```

```python
X.flatten() # outplace
```

```python
X = X.flatten()
```

```python
X[[0, 2, 4]] #0, 2, 4番目の要素を取得
```

```python
X[np.array([0, 2, 4])] #0, 2, 4番目の要素を取得
```

```python
X > 15
```

```python
(X > 15).dtype
```

```python
X[X>15]
```

## matlabの復習

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 6, 0.1) # 0~6まで0.1刻みで作成
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.show()
```

```python
y2 = np.cos(x)
```

```python
plt.plot(x, y, label="sin")
plt.plot(x, y2, label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title('sin & cos')
plt.legend()
plt.show()
```
